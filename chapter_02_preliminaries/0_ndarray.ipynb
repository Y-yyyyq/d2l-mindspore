幸运的是，执行原地操作非常简单。 我们可以使用切片表示法将操作的结果分配给先前分配的数组，例如Y[:] = <expression>。 为了说明这一点，我们首先创建一个新的矩阵Z，其形状与另一个Y相同， 使用zeros_like来分配一个全的块。

```python
Z = ops.zeros_like(Y)
print('id(Z):', id(Z))
Z[:] = X + Y
print('id(Z):', id(Z))
```

如果在后续计算中没有重复使用X， 我们可以使用X[:] = X + Y来减少操作的内存开销。

```python
before = id(X)
X[:] = X + Y
id(X) == before
```
